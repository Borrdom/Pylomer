# Water sorption isotherm of an active pharmaceutical ingredient

Calculation of the water sorption isotherm of indomethacin (see the example in Figure 1 from https://doi.org/10.1021/acs.molpharmaceut.1c00984). 

```python
import numpy as np
from Pylomer import lnai
import matplotlib.pyplot as plt


Mi=np.asarray([18.015,357.787])
rho0i=np.asarray([997.,1320.])
chi=np.asarray([2.7067])
w1=np.linspace(0.0001,0.05,50)
w2=1-w1

wi=np.stack((w1,w2))

lnaivec=np.asarray([lnai(wi[:,i],rho0i,Mi,chi=chi) for i,val in enumerate(wi[0,:])]).T
RH=np.exp(lnaivec[0,:])

font = {'weight' : 'normal',
        'size'   : 12}
plt.rc('font', **font)

color = "#FF8500"

fig, ax = plt.subplots(figsize = (4, 3), dpi = 300)
fig.subplots_adjust(hspace=0.5, wspace=0.3)
wexp=np.asarray([0.088,0.387,0.704,1.13,1.64,2.43])
RHexp=np.asarray([10,30,45,60,75,90])
ax.plot(RH*100,w1*100,"--",color = color ,  linewidth = 2.0)
ax.plot(RHexp,wexp, 'o', markersize=12, markerfacecolor=color, markeredgewidth=1.5, markeredgecolor="k")
ax.set_xlabel("RH/%")
ax.set_ylabel("$w_w$/%")
plt.xlim([0,100])
plt.ylim([0,5])
plt.show()
```
