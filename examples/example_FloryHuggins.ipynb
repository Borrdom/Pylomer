# Water sorption Isotherm of an active pharmaceutical ingredient

Calculation of the w(see the example in Table 2 from https://doi.org/10.1021/acs.molpharmaceut.1c00984). Furthermore, this book explains statements which can be made from its values.

```python
import numpy as np
from Pylomer import lnai
import matplotlib.pyplot as plt


Mi=np.asarray([18.015,721])
rho0i=np.asarray([997.,1150.])
chi=np.asarray([4.7])
w1=np.linspace(0.0001,0.05,100)
w2=1-w1

wi=np.stack((w1,w2))

lnaivec=np.asarray([lnai(wi[:,i],Mi,chi=chi) for i,val in enumerate(wi[0,:])]).T
RH=np.exp(lnaivec[0,:])

font = {'weight' : 'normal',
        'size'   : 12}
plt.rc('font', **font)

color = "#000000"

fig, ax = plt.subplots(figsize = (4, 3), dpi = 300)
fig.subplots_adjust(hspace=0.5, wspace=0.3)
wexp=np.asarray([0.123,0.652,1.3,2.1,3,4.34])
RHexp=np.asarray([10,30,45,60,75,90])
ax.plot(RH*100,w1*100,"--",color = color ,  linewidth = 2.0)
ax.plot(RHexp,wexp, 'o', markersize=12, markerfacecolor='w', markeredgewidth=1.5, markeredgecolor=color)
ax.set_xlabel("RH/%")
ax.set_ylabel("$w_w$/%")
plt.show()
```
