# Water sorption kinetics in an active pharmaceutical ingredient

Calculation of the water-sorption kinetics in indomethacin (see the example in Figure 2 from https://doi.org/10.1021/acs.molpharmaceut.1c00984). 

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import array
from Pylomer import diffuse,crank
t=np.linspace(0,120,100)*60

Lvec=np.asarray([8.67E-13])
length=9.93E-6
ww0=1.10671/100
ww8=1.6374/100
wi0=np.asarray([ww0,1-ww0])
wi8=np.asarray([ww8,1-ww8])
mob=np.asarray([True,False])
```

We model the downside triangles: the step from 0.6 to 0.75 RH

```python

witz=diffuse(t,length,Lvec,wi0,wi8,mob)
wit=np.average(witz,axis=1)
wt=crank(t,Lvec[0],length,ww0,ww8)
font = {'weight' : 'normal',
        'size'   : 12}
plt.rc('font', **font)

color = "#FF8500"
wexp=array([1.10671,1.24201, 1.37726, 1.50304, 1.58311, 1.61596, 1.62772, 1.62671,
       1.62944, 1.63171, 1.63199, 1.63472, 1.63834, 1.64072, 1.63991,
       1.6374 ])
texp=array([  0, 1.5932 ,   3.98592,   8.32066,  15.27155,  24.26213,  33.73908,
        43.92737,  52.89087,  63.27516,  72.8396 ,  82.7068 ,  90.38047,
        98.85767, 109.32944, 119.4    ])
fig, ax = plt.subplots(figsize = (4, 3), dpi = 300)
fig.subplots_adjust(hspace=0.5, wspace=0.3)
ax.plot(t/60,wit[:,0]*100,color = color ,  linewidth = 2.0)
ax.plot(t/60,wt*100,color = "#FF8500" ,  linewidth = 1.0)
ax.plot(texp,wexp, 'o', markersize=12, markerfacecolor=color, markeredgewidth=1.5, markeredgecolor="k")
ax.set_xlabel("t/min")
ax.set_ylabel("$w_w$/%")
```
